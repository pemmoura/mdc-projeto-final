```python
models = ["bert-tiny", "bert-small", "bert-base", "roberta", "distilroberta", "electra"]
```

```python
import matplotlib.pyplot as plt
print(plt.style.available)
def plot_model_comparison(models, balanced_accuracy, metric='Acurácia Balanceada', limit=1, labels_45=False):
    plt.style.use('seaborn-v0_8-white')
    plt.figure(figsize=(10, 6))
    plt.bar(models, balanced_accuracy, color='skyblue', edgecolor='black')
    plt.xlabel('Modelos', fontsize=12)
    plt.ylabel(metric, fontsize=12)
    plt.title('Comparação de '+metric+' entre modelos', fontsize=14)
    plt.ylim(0, limit)  # Limite do eixo y para refletir acurácias como proporções
    for i, acc in enumerate(balanced_accuracy):
        plt.text(i, acc + 0.02, f"{acc:.2f}", ha='center', fontsize=10)
    if labels_45:
        plt.xticks(rotation=45, ha='right', fontsize=10)  # Rótulos em 45 graus
        plt.tight_layout()  # Ajusta para evitar corte de elementos
    plt.show()
```

```python
balanced_accuracy_critic = [0.69, 0.77, 0.74, 0.7, 0.82, 0.85]
plot_model_comparison(models, balanced_accuracy_critic)
```

```python
balanced_accuracy_audience = [0.29, 0.32, 0.32, 0.30, 0.33, 0.34]
plot_model_comparison(models, balanced_accuracy_audience, limit=0.4)
```

```python
mae_audience = [1.24, 1.11, 1.09, 1.24, 1.02, 0.92]
mae_audience =  [m / 2 for m in mae_audience]
plot_model_comparison(models, mae_audience, limit=0.7, metric='Erro Médio Absoluto')
```

```python
models_trad = ["SVM", "MultinomialNB", "Ridge Classifier", "Neural Network", "SVC", "Stochastic", "Gradient Descent", "AdaBoostClassifier", "Gradient Boosting", "Random Forest", "K-Nearest Neighbors", "CatBoost", "XGBoost", "Radius Neighbors"]

balanced_accuracy_trad = [0.71, 0.61, 0.61, 0.60, 0.60, 0.60, 0.60, 0.60, 0.60, 0.56, 0.56, 0.55, 0.53, 0.50]
plot_model_comparison(models_trad, balanced_accuracy_trad, limit=0.77, labels_45=True)
```
