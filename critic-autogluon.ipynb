```python
from autogluon.tabular import TabularDataset, TabularPredictor

data_root = './'
train_data = TabularDataset(data_root + 'train_critic.csv')
test_data = TabularDataset(data_root + 'test_critic.csv')

predictor = TabularPredictor(label='Sentiment').fit(train_data=train_data)
predictions = predictor.predict(test_data)
```

```python
predictor.leaderboard(test_data)
```

```python
predictor.evaluate(test_data, silent=True)
```
